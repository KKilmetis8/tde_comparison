```python
L_n = np.loadtxt('Lum_n_m4.txt')
array = L_n[0]

L233 = L_n[1]
L263 = L_n[2]
L262 = L_n[3]

i233 = np.log(10)*np.trapz(10**(array)*L233, array)
i263 = np.log(10)*np.trapz(10**(array)*L263,array)
i262 = np.log(10)*np.trapz(10**(array)*L262,array)


print('233: ', i233) 
print('263: ', i263)
```
